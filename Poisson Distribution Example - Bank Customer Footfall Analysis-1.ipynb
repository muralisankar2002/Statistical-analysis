# Poisson distribution Example

In a manufacturing company  on average in production 6 items are defective per day.
a)What is the probability that in production exactly three items are defective per day? 
b)What is the probability that in production more than four items per day will be defective?
c)Plot the graph of probability mass function?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# poisson parameter lambda.
lambda_value = 6
```

```python
n=np.arange(0,20)
```

```python
poisson_values = poisson.pmf(n,lambda_value)
```

```python
poisson_values
```

```python
#The probability exactly three items are defective per day.
poisson_values[3]
```

```python
#To find the probability that more than four customers will arrrive in a given minute.
total = 0
for i in range(0,4):
    total += poisson_values[i]
p = 1-total
print("The probability of more than four items produced are defective is %f" %p)
```

###  Plot of probability mass function curve

```python
plt.plot(n,poisson_values,'o-')
plt.title('Poisson: $\lambda$ = %i ' % lambda_value)
plt.xlabel('Number of Customers Arriving in a minute')
plt.ylabel('Probability of Number of Customers arriving in a minute')
plt.show()
```
